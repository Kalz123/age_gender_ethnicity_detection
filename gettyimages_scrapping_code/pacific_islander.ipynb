```python
# import libraries
import requests
from bs4 import BeautifulSoup
import urllib.request
import progressbar
phrase = input("Phrase: ")

count = 0
for page in range(1,101):
    get_soup = requests.get(
    f"https://www.gettyimages.com/photos/human-face?compositions=headshot&ethnicity=pacificislander&family=creative&numberofpeople=one&page={page}&phrase={phrase}%20face&sort=best")
    soup = BeautifulSoup(get_soup.text, "lxml")
    pics = soup.findAll("img", {"class": "gallery-asset__thumb gallery-mosaic-asset__thumb"})
    for pic in progressbar.progressbar(pics, prefix=f"page: {page} [{phrase}]"):
        try:
            urllib.request.urlretrieve(pic.get("src"), filename = f"/Users/kalz/Desktop/CAPSTONE/ethnicity/pacific_islander/{count}.jpg")
        except:
            pass
        count += 1
```
